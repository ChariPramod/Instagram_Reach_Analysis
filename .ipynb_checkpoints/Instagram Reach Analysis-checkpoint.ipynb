```python
import pandas as pd
import wordcloud
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud, STOPWORDS,ImageColorGenerator
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveRegressor
```

```python
data = pd.read_csv("InstagramData.csv",encoding = 'latin1')
print(data.head())
```

```python
#Checking for null values
data.isnull().sum()
```

```python
#Luckily no null values but still writing a code to drop null values
data = data.dropna()
```

```python
data.info
```

```python
# Analysing the reach of Instagram Posts.
# Looking at impression recieved from home page
plt.figure(figsize = (10,8))
plt.style.use('fivethirtyeight')
plt.title("Distribution of Impressions From Home")
sns.distplot(data['From Home'])
plt.show()

#Impressions recived from home section shows how many posts reach the instagram users followers
#From impression we can analyse that the user dosent get many followers daily from the home page 
```

```python
#Distribution of Impression from hashtags
plt.figure(figsize = (10,8))
plt.title("Distribution of Impressions From Hashtags")
sns.distplot(data['From Hashtags'])
plt.show()

#Hashtags are used to categorize posts on Instagram so that ours posts can reach more audience
#Impressions from hashtags shows that not all the posts can be reached through hashtags
```
